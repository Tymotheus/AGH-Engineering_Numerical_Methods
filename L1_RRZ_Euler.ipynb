<h2> 1. Metoda jawna Eulera </h2>

```python
#Rozwiazanie metodą jawną Eulera dla trzech kroków czasowych
#1 delta t = 0.01
d_t1 = 0.01
t_tab1 = np.arange(t_min, t_max, d_t1)
y_tab1 = np.zeros(t_tab1.size )
real_tab1 = np.zeros(t_tab1.size)
error_tab1 = np.zeros(t_tab1.size)
y_tab1[0] = y_0
real_tab1[0] = y_0
with open("Euler.txt", "w") as file:
    file.write("Delta t = 0.01\n")
    file.write("t   y   error\n")
    for i in range(1, 500):
        y_tab1[i] = y_tab1[i-1] + d_t1 * lambd * y_tab1[i-1]
        real_tab1[i] = exp(lambd * t_tab1[i])
        error_tab1[i] = y_tab1[i] - real_tab1[i]
        file.write('' + str(t_tab1[i]) +' ' + str(y_tab1[i]) + ' ' + str(error_tab1[i]) + '\n')
```

<h2> 4. Obwód RLC i RRZ 2 rzędu </h2>

```python
#Initialising parameters
#time in seconds
dt = 0.0001
#resistance of the resistor in ohms
R = 100
#inductance of the coil in henries
L = 0.1
#capacitance of the capacitor in farads
C = 0.001
#frequency in t0, in Hertz
omega0 = 1/sqrt(L*C)
#period in t0 in seconds
T0 = 2*pi / omega0
#time range
t_min = 0
t_max = 4*T0
#electric charge in t_0
Q_0 = 0
#electric current in t_0
I_0 = 0
```

```python
def voltage(time, freq):
    """Voltage function in a given time for a given frequency"""
    return 10 * sin(time * freq)
```

```python
t_tab = np.arange(t_min, t_max, dt)
v_real = np.zeros(t_tab.size)
for i in range(0, t_tab.size):
    v_real[i] = voltage(t_tab[i], omega0)
```

```python
#1 omega v = 0.5 omega 0
omegav_1 = 0.5 * omega0
Q_tab1 = np.zeros(t_tab.size)
I_tab1 = np.zeros(t_tab.size)
V_tab1 = np.zeros(t_tab.size)
Q_tab1[0] = Q_0
I_tab1[0] = I_0
V_tab1[0] = 0

with open('RRZ.txt', 'w') as file:
    file.write("Omega_v = 0.5 * omega_0 \n")
    file.write("t   Q   I\n")
    for i in range(1, t_tab.size):
        kQ_1 = I_tab1[i-1]
        kI_1 = voltage(t_tab[i-1], omegav_1)/L  - 1/(L*C) * Q_tab1[i-1] - (R/L) * I_tab1[i-1] 
        kQ_2 = I_tab1[i-1] + (dt/2)*kI_1
        kI_2 = voltage(t_tab[i-1] + dt/2, omegav_1)/L - 1/(L*C) * (Q_tab1[i-1] + dt/2 * kQ_1) - R/L * (I_tab1[i-1] + dt/2 * kI_1)
        kQ_3 = I_tab1[i-1] + dt/2 * kI_2
        kI_3 = voltage(t_tab[i-1] + (dt/2), omegav_1)/L - 1/(L*C) * (Q_tab1[i-1] + dt/2 * kQ_2) - R/L * (I_tab1[i-1] * dt/2 * kI_2)
        kQ_4 = I_tab1[i-1] + dt * kI_3
        kI_4 = voltage(t_tab[i-1] + dt, omegav_1)/L - 1/(L*C) * (Q_tab1[i-1] + dt * kQ_3) - R/L * (I_tab1[i-1] + dt *kI_3)  
        Q_tab1[i] = Q_tab1[i-1] + dt/6 * (kQ_1 + 2*kQ_2 + 2*kQ_3 + kQ_4)
        I_tab1[i] = I_tab1[i-1] + dt/6 * (kI_1 + 2*kI_2 + 2*kI_3 + kI_4)
        file.write('' + str(t_tab[i]) + ' ' + str(Q_tab1[i]) + ' ' + str(I_tab1[i]) + '\n')
        
```

```python
#2
omegav_2 = 0.8 * omega0
Q_tab2 = np.zeros(t_tab.size)
I_tab2 = np.zeros(t_tab.size)
V_tab2 = np.zeros(t_tab.size)
Q_tab2[0] = Q_0
I_tab2[0] = I_0
V_tab2[0] = 0

with open('RRZ.txt', 'a') as file:
    file.write("\nOmega_v = 0.8 * omega_0 \n")
    file.write("t   Q   I\n")
    for i in range(1, t_tab.size):
        kQ_1 = I_tab2[i-1]
        kI_1 = voltage(t_tab[i-1], omegav_2)/L  - 1/(L*C) * Q_tab2[i-1] - (R/L) * I_tab2[i-1] 
        kQ_2 = I_tab2[i-1] + (dt/2)*kI_1
        kI_2 = voltage(t_tab[i-1] + dt/2, omegav_2)/L - 1/(L*C) * (Q_tab2[i-1] + dt/2 * kQ_1) - R/L * (I_tab2[i-1] + dt/2 * kI_1)
        kQ_3 = I_tab2[i-1] + dt/2 * kI_2
        kI_3 = voltage(t_tab[i-1] + (dt/2), omegav_2)/L - 1/(L*C) * (Q_tab2[i-1] + dt/2 * kQ_2) - R/L * (I_tab2[i-1] * dt/2 * kI_2)
        kQ_4 = I_tab2[i-1] + dt * kI_3
        kI_4 = voltage(t_tab[i-1] + dt, omegav_2)/L - 1/(L*C) * (Q_tab2[i-1] + dt * kQ_3) - R/L * (I_tab2[i-1] + dt *kI_3)  
        Q_tab2[i] = Q_tab2[i-1] + dt/6 * (kQ_1 + 2*kQ_2 + 2*kQ_3 + kQ_4)
        I_tab2[i] = I_tab2[i-1] + dt/6 * (kI_1 + 2*kI_2 + 2*kI_3 + kI_4)
        file.write('' + str(t_tab[i]) + ' ' + str(Q_tab2[i]) + ' ' + str(I_tab2[i]) + '\n')
```
